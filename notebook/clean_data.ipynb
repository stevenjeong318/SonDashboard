```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw_cleaned.csv")
```

```python
df.head()
```

```python
#rename the unnamed & confusing column names
df = df.rename(columns={"Competition":"Season","Unnamed: 1": "League", "For": "Played for"})
df.head()
```

```python
#replace values in season columns (get rid of word 'season' and write in 20xx/20xx form)
df['Season'].replace(to_replace=r'Season (\d{2})/(\d{2})', value=r'20\1/20\2', regex=True, inplace=True)
df.head()
```

```python
#fill 'played for' column (team)
def calculate_played(season):
    if season == '2008/2009':
        return 'Hamburger SV U-17'
    elif season == '2009/2010':
        return 'Hamburger SV U-19'
    elif (season == '2010/2011') or (season == '2011/2012') or (season == '2012/2013'):
        return 'Hamburger SV'
    elif (season == '2013/2014') or (season == '2014/2015'):
        return 'Bayer 04 Leverkusen'
    else:
        return 'Tottenham Hotspur F.C.'
        
df['Played for'] = df['Season'].apply(calculate_played)
```

```python
#remove records from U-17 and U-19 (not pro contract & missing values)
seasons_to_remove = ['2008/2009', '2009/2010']
df = df[~df['Season'].isin(seasons_to_remove)]
df.reset_index(drop=True, inplace=True)
```
